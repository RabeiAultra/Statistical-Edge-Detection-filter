```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.filters.rank import equalize
from skimage.morphology import disk
from skimage.restoration import (denoise_tv_chambolle, denoise_bilateral,
                                 denoise_wavelet, estimate_sigma)
img = cv2.imread('images/results/Stastical_Filter_MP_K_7.png',0)      
#img = cv2.imread('images/StasticalFilter.png',0)   
#blur = cv2.bilateralFilter(img,9,50,50)
#median = cv2.medianBlur(img,25) 
img2 = cv2.equalizeHist(img)    
#img3 =denoise_tv_chambolle(img2, weight=0.2, multichannel=True)



plt.figure(figsize=(40, 40))
plt.subplot(3,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,2),plt.imshow(img2,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,3),plt.imshow(img2,cmap = 'gray')
plt.title('After'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
from nonmax_suppression import maximum
from __future__ import division
from gaussian_filter import gaussian
from gradient import gradient
from numpy import array, zeros
from PIL import Image
from double_thresholding import thresholding

img2= cv2.imread('images/results/sobel2.jpg',0)   
img2 = cv2.bilateralFilter(img2,7,75,75)
th2 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,9,2)

gim = gaussian(th2)
grim,gphase= gradient(img2)
maxim = maximum(gim, gphase)
gmax = thresholding(maxim)



plt.figure(figsize=(100, 100))
plt.subplot(3,2,1),plt.imshow(img2,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,2),plt.imshow(gim,cmap = 'gray')
plt.title('Gaussian'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,3),plt.imshow(grim,cmap = 'gray')
plt.title('Gradient'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,4),plt.imshow(maxim,cmap = 'gray')
plt.title('Non_maximum Supresstion'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,5),plt.imshow(gmax[0],cmap = 'gray')
plt.title('Double Thresholding'), plt.xticks([]), plt.yticks([])

plt.show()
```
